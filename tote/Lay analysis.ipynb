# Analysis on profitability of laying bets

```python
import pymysql
import sqlalchemy
import pandas as pd
import numpy as np
import sys
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
connect_string = 'mysql+pymysql://root:angunix1@localhost/horses'
sql_engine = sqlalchemy.create_engine(connect_string)
horses_data = pd.read_sql('''
                          SELECT DISTINCT h.race_date, h.race_id, h.course, h.surface, h.going, h.has_handicap, h.runners,
                                  h.race_class, h.ride_status, h.finish_position,
                                  case when h.finish_position=1 THEN 'back_win' else 'lay_win' end as outcome,
                                  h.betting_odds
                          FROM horses_data h
                          LEFT JOIN races_data r ON h.race_id = r.race_id 
                          WHERE h.race_date > '2020-01-01'
                          AND country IN ('Scotland', 'England', 'Ireland', 'Wales', 'Eire', 'Northern Ireland',
                                          'United Kindgom')
                          ''',
                          con=sql_engine)
```

```python
def odds_parser(odds_string):
    try:
        odds_split = odds_string.split('/')
        decimal_odds = (int(odds_split[0])+int(odds_split[1]))/int(odds_split[1])
        return decimal_odds
    except:
        return 1
```

```python
horses_data['decimal_odds'] = horses_data['betting_odds'].apply(odds_parser)
```

```python
low_lay_sp_ratios = [
#2.92,
1.11,
1.30,
1.50,
1.31,
1.83,
0.77,
#3.00,
1.00,
1.24,
1.54,
1.68,
#2.00,
0.85,
1.22,
0.88,
1.15,
0.85,
0.71,
0.81,
1.30,
1.54,
1.50,
0.81,
1.00,
1.81,
0.91,
1.38,
0.51,
1.08,
1.24,
1.31,
1.20,
0.72,
1.67
]

high_lay_sp_ratios = [
2.92,
1.11,
1.30,
1.50,
1.31,
1.83,
0.77,
3.00,
1.00,
1.24,
1.54,
1.68,
2.00,
0.85,
1.22,
0.88,
1.15,
0.85,
0.71,
0.81,
1.30,
1.54,
1.50,
0.81,
1.00,
1.81,
0.91,
1.38,
0.51,
1.08,
1.24,
1.31,
1.20,
0.72,
1.67
]
```

```python
lay_margin_pc = 0.05
fixed_margin = False

if fixed_margin:
    horses_data['lay_odds'] = horses_data['decimal_odds']*(1+lay_margin_pc)
else:
    horses_data['lay_sp_ratio'] = np.random.choice(low_lay_sp_ratios, size=len(horses_data), replace=True)
    horses_data.loc[horses_data['decimal_odds']>=7, 'lay_sp_ratio'] = np.random.choice(
        high_lay_sp_ratios, size=sum(horses_data['decimal_odds']>=7), replace=True)
    horses_data['lay_odds'] = horses_data['decimal_odds']*horses_data['lay_sp_ratio']
```

```python
lay_stake = 10
horses_data['lay_stake'] = lay_stake
horses_data['lay_liab'] = (horses_data['lay_odds']-1)*lay_stake
```

```python
exch_comm = 0.02
horses_data['lay_win'] = (horses_data['outcome']=='lay_win')
horses_data['lay_p_l'] = (
    horses_data['lay_win']*horses_data['lay_stake']*(1-exch_comm) -
    (1-horses_data['lay_win'])*horses_data['lay_liab'])
```

```python
band_size = 0.05
horses_data['p_lay'] = (1-1/horses_data['lay_odds'])
horses_data['p_lay_band_min'] = ((horses_data['p_lay']/band_size).apply(np.floor)*band_size).round(2)
horses_data['p_lay_band_max'] = (horses_data['p_lay_band_min']+band_size).round(2)
horses_data['p_lay_band'] = (
    horses_data['p_lay_band_min'].astype(str)+' - '+horses_data['p_lay_band_max'].astype(str))
```

```python
p_and_l = horses_data.groupby('p_lay_band').agg({
    'race_id': [lambda x: len(set(x)), len],
    'outcome': lambda x: sum(np.array(x)=='lay_win'),
    'lay_liab': sum,
    'lay_p_l': sum,
    'race_date': lambda x: len(set(x))
}).reset_index()
p_and_l.columns = ['p_lay_band', 'races', 'bets_laid', 'lay_wins', 'lay_liab', 'lay_p_l', 'days_wagered']

p_and_l['roaw'] = 100*(p_and_l['lay_p_l'] / p_and_l['lay_liab'])

p_and_l['lay_liab_per_day'] = p_and_l['lay_liab']/p_and_l['days_wagered']
p_and_l['roi'] = 100*(p_and_l['lay_p_l'] / p_and_l['lay_liab_per_day'])

p_and_l
```

#### Daily accumulation

```python
daily_p_and_l = horses_data.groupby(['race_date', 'p_lay_band']).agg({
    'race_id': [lambda x: len(set(x)), len],
    'outcome': lambda x: sum(np.array(x)=='lay_win'),
    'lay_liab': sum,
    'lay_p_l': sum,
    'race_date': lambda x: len(set(x))
}).reset_index()
daily_p_and_l.columns = [
    'race_date', 'p_lay_band', 'races', 'bets_laid', 'lay_wins', 'lay_liab', 'lay_p_l', 'days_wagered']

daily_p_and_l['roaw'] = 100*(daily_p_and_l['lay_p_l'] / daily_p_and_l['lay_liab'])

daily_p_and_l['lay_liab_per_day'] = daily_p_and_l['lay_liab']/daily_p_and_l['days_wagered']
daily_p_and_l['roi'] = 100*(daily_p_and_l['lay_p_l'] / daily_p_and_l['lay_liab_per_day'])

daily_p_and_l
```

```python
daily_p_and_l = daily_p_and_l.sort_values('race_date')
daily_p_and_l['lay_p_l_cumulative'] = daily_p_and_l.groupby('p_lay_band')['lay_p_l'].cumsum()

daily_p_and_l['race_datetime'] = pd.to_datetime(daily_p_and_l['race_date'])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 10))

lay_bands_to_plot = ['0.75 - 0.8', '0.8 - 0.85', '0.85 - 0.9', '0.9 - 0.95', '0.95 - 1.0']

sns.lineplot(data=daily_p_and_l[daily_p_and_l['p_lay_band'].isin(lay_bands_to_plot)],
             x='race_datetime', y='lay_p_l_cumulative', hue='p_lay_band')
plt.xticks(rotation=30)
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 10))

lay_band = '0.95 - 1.0'

sns.distplot(daily_p_and_l.loc[daily_p_and_l['p_lay_band']==lay_band, 'lay_p_l'], kde=False)
```

## How many bets needed to show lift

```python
bookie_odds = 1.05
actual_odds = 1.03
bookie_prob = 1/bookie_odds
actual_prob = 1/actual_odds
events = 1000
n_bets = 100
bootstraps = 1000

np.random.seed(123)
actual_outcomes = np.random.binomial(n=1, p=actual_prob, size=events)

bootstraps = [sum(np.random.choice(actual_outcomes, size=n_bets))*bookie_odds - n_bets
              for b in tqdm(range(bootstraps))]
```

```python
bookie_back_odds = 1/(1-bookie_prob)
bookie_back_odds
```

```python
actual_back_odds = 1/(1-actual_prob)
actual_back_odds
```
